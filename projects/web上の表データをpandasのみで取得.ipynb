```python
import pandas as pd
```

https://info.finance.yahoo.co.jp/ranking/

```python
url = 'https://info.finance.yahoo.co.jp/ranking/'
dfs = pd.read_html(url)
```

```python
len(dfs)
```

```python
df = dfs[0]
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.columns = ['順位', 'コード', '市場', '名称', '日付', '取引値', '前日比', '増加値', '出来高', '掲示板']
df.head()
```

```python
df.tail()
```

```python
df.iloc[:-1, :].tail()
```

```python
df.index[-1]
```

```python
df.drop(df.index[-1], inplace=True)
df.tail()
```

```python
df['順位'].dtype
```

```python
df = df.astype({'順位': int})
```

```python
df.info()
```

```python
df['増加値'] = [dt.replace('+', '') for dt in df['増加値'].tolist()]
```

```python
df['増加値'] = [dt.replace(',', '') for dt in df['増加値'].tolist()]
```

```python
df.head()
```

```python
df = df.astype({
    '順位': int, 
    'コード': int, 
    '取引値': int, 
    '増加値': int, 
    '出来高': int, 
    })
```

```python
df.dtypes
```

```python
df['前日比'] = [dt.replace('+', '').replace('%', '') for dt in df['前日比'].tolist()]
```

```python
df.head()
```

```python
df = df.astype({
    '前日比': float
    })
```

```python
df.dtypes
```

```python
import datetime
```

```python
df['日付'] = [datetime.datetime.strptime('2021'+dt, '%Y%m/%d') for dt in df['日付'].tolist()]
```
